```python
def square(n):
    squares = []
    for i in range(n):
        squares.append(i**2)
    return squares
print(square(10))
```

```python
class Observation(): #"object" and "class" are interchangeable
    def __init__(self, data): #Special method called constructor 
        self.data = data #attribute
    def average(self): #method
        dsum = 0 
        for i,d in enumerate(self.data):
            dsum += d
        average=dsum/(i+1)
        return average
    
```

```python
obs1 = Observation([0,1,2])
obs2 = Observation([4,5,6])
```

```python
print("Avg 1 = {:e}; Avg 2= {:e}".format(obs1.average(), obs2.average()))
print("Type of Avg 1 = {:}; Type of Avg 2 = {:}".format(type(obs1), type(obs2)))
```

```python
print(obs1.data)
print(obs2.data)
```

```python
class TimeSeries(Observation): #inherits all the methods and attributes of Observations
    def __init__(self,time,data):
        self.time = time
        Observation.__init__(self, data) #this calls the class Observation
        if len(self.time) != len(self.data):
            raise ValueError("Time and data must have same length")
    def stop_time(self):
        return self.time[-1] # unclear why you would want this
```

```python
tobs = TimeSeries([0,1,2],[3,4,5])
print(tobs)
print("Stop time = {:e}".format(tobs.stop_time()))
print("tobs average = {:e}".format(tobs.average()))
```

```python
x = np.linspace(0,2*np.pi, 1000)
y_theory = np.sinc(x)
y = y_theory + np.random.rand(1000)
```

```python
fig = plt.figure(figsize=(8,8)) # create a figure object
ax_data = fig.add_axes([0.1,0.4,0.8,0.8])
ax_residual = fig.add_axes([0.1,0.1,0.8,0.3])

#this is one axis
ax_data.plot(x,y,label='sinc(x)') #plot is a method of 
ax_data.legend()
ax_data.set_ylabel('f(x)') # labels are attribues

ax_residual.plot(x,y-y_theory, label='residual')
ax_residual.legend()
ax_residual.set_xlabel('x')
ax_residual.set_ylabel('residual')
```
